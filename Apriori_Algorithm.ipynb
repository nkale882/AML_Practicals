```python
#pandas for handling data (like reading Excel or CSV files),
#apriori from the apyori library for association rule mining.

import pandas as pd
from apyori import apriori
```

```python
"""Loads your Excel file named apriori.xlsx
Doesn’t treat the first row as column headers (thanks to header=None)
Stores the content in a DataFrame df"""

df=pd.read_excel('apriori.xlsx',header=None)
```

```python
#Running df.head() will show you the first 5 rows of your Excel data.

df.head()
```

```python
#Running df.shape will return a tuple showing the number of rows and columns in your DataFrame.

df.shape
```

```python
transaction = list()
for i in range(df.shape[0]):
    transaction.append([str(df.values[i, j]) for j in range(df.shape[1])])
print("List of items in df")
print(transaction)
```

```python
#buid the Apriori algorithm
final_rule = apriori(transaction,min_support=0.4,min_confidence=0.8,min_lift=1.2,min_length=2)
print("Building the association rule")
final_results= list(final_rule)

#Print the number of rules
print("Number of rules")
print(len(final_results))
```

```python

#Printin all the rules
for i in range(len(final_results)):
  print('*****************************************')
  print('rule'+str(i))
  print(final_results[i])
  print('*****************************************')
```

```python
rule_number = 1

for rule in final_results:
    for stat in rule.ordered_statistics:
        if len(stat.items_base) > 0:
            base = ', '.join(stat.items_base)
            add = ', '.join(stat.items_add)
            print(f"Rule #{rule_number}: {base} -> {add}")
            print(f"Support: {rule.support:.2f}")
            print(f"Confidence: {stat.confidence:.2f}")
            print(f"Lift: {stat.lift:.2f}")
            print("-----")
            rule_number += 1
```
